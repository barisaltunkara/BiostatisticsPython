```python
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt
from scipy import stats
```

```python
x = np.array([3265, 3260, 3245, 3484, 4146,
              3323, 3649, 3200, 3031, 2069,
              2581, 2841, 3609, 2838, 3541,
              2759, 3248, 3314, 3101, 2834, 3541])
```

```python
print("Sum of the list:\t\t", np.sum(x))
print("Arithmetic Mean of the list:\t", np.mean(x))
print("Geometric Mean of the list:\t", stats.gmean(x))
print("Median of the list:\t\t", np.median(x))
print("Mode of the list:\t\t", stats.mode(x)[0])
print("Range of the list:\t\t", np.max(x)-np.min(x))
```

```python
x = np.append(x, 100000)
```

```python
print("Sum of the list:\t\t", np.sum(x))
print("Arithmetic Mean of the list:\t", np.mean(x))
print("Geometric Mean of the list:\t", stats.gmean(x))
print("Median of the list:\t\t", np.median(x))
print("Mode of the list:\t\t", stats.mode(x)[0])
print("Range of the list:\t\t", np.max(x)-np.min(x))
```

### Deviance

```python
np.sum(x-np.mean(x))/len(x)
```

### Mean Deviation

```python
np.sum(np.abs(x-np.mean(x)))/len(x)
```

### Sample Variance

```python
sam_var = np.sum((x-np.mean(x))**2)/(len(x)-1)
sam_var
```

### Sample Standard Deviation

```python
sam_st = np.sqrt(np.sum((x-np.mean(x))**2)/(len(x)-1))
sam_st
```

### Coefficient of Variation

```python
CV = 100*(sam_st/np.mean(x))
CV
```

### Quantiles

```python
Q3 = np.quantile(x, 0.75)
med = np.quantile(x, 0.5)
Q1 = np.quantile(x, 0.25)
Q3, med, Q1
```

### IQR - Interquartile Range

```python
IQR = Q3 - Q1
IQR
```
